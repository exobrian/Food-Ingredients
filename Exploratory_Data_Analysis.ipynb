## Removing Zero Calorie Foods from our table

```python
# Removing zero calorie food
caloric_foods = rawDataFrame.drop(labels=np.where(rawDataFrame['Energy_kcal']==0)[0], 
                                  axis=0, 
                                  inplace=False)
caloric_foods.head()
```

```python
caloric_foods.info()
```

## Interesting note about the graph below: 

The total Calories of each datapoint is not isolated to the macronutrient. A Calorie is made up of a combination of macros which contribute to the overall caloric count. It is typically assumed that a unit of Protein, Carbohydrate, and Fat contributes roughly 4, 4, and 9 units of Calories respectively.

```python
caloric_foods_unique = caloric_foods[caloric_foods.duplicated(subset=['Descrip'], keep='first') == False]
print(str(caloric_foods.shape[0] - caloric_foods_unique.shape[0]) + " duplicate food items have been removed")

kcals = caloric_foods_unique['Energy_kcal']
proteins = caloric_foods_unique['Protein_g']
carbs = caloric_foods_unique['Carb_g']
fats = caloric_foods_unique['Fat_g']

plt.scatter(y=kcals, x=proteins, c='red', s=1, label="Protein (g)")
plt.scatter(y=kcals, x=carbs, c='blue', s=1, label="Carbohydrates (g)")
plt.scatter(y=kcals, x=fats, c='green', s=1, label="Fats (g)")

plt.title("Calories vs Macro Nutrients", fontsize=15)
plt.xlabel("Macronutrients", fontsize=10)
plt.ylabel("Calories per serving", fontsize=10)
plt.legend()
plt.show()
```
